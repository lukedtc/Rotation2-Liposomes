```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import trapz
```

```python
data = pd.read_csv('/home/heirene/Rotation2-liposomes/COMSOL project/mass_conserv_3d.csv')
```
